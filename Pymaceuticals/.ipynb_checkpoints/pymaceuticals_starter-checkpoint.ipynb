```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
import matplotlib.ticker as mtick

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"
```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data
drug_data = pd.read_csv(mouse_drug_data_to_load)
trial_data = pd.read_csv(clinical_trial_data_to_load)

# check the count of Mouse IDs on both dataframes to decide on type of join
# in this case both had same amount of entries, so the outer or left or right will produce same result
# drug_data['Mouse ID'].value_counts()
# trial_data['Mouse ID'].value_counts()

# Combine the data into a single dataset
mouse_trial_data = pd.merge(trial_data, drug_data, how='left', on='Mouse ID')
```

```python
# Display the data table for preview
mouse_trial_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# uses reset_index() method to not use group by columns as index.
tumor_group = mouse_trial_data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean().reset_index()

# tumor_group = tumor_data.groupby(['Drug','Timepoint'], as_index=False)['Tumor Volume (mm3)'].mean() #alternate method using as_index=False argument

# Convert to DataFrame
tumor_df = pd.DataFrame(tumor_group)

# Preview DataFrame
tumor_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_err_group = mouse_trial_data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].sem().reset_index()

# Convert to DataFrame

tumor_std_err = pd.DataFrame(tumor_err_group)

# Preview DataFrame

tumor_std_err.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_pivot = tumor_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

tumor_err = tumor_std_err.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
tumor_pivot
```

```python
tumor_err
```

```python
# Generate the Plot (with Error Bars)
# Setting up the plot

fig, ax = plt.subplots()

capomulin, = plt.plot(tumor_pivot.index, tumor_pivot.Capomulin, marker='o', color="red", linewidth=0.5, label="Capomulin")
infubinol, = plt.plot(tumor_pivot.index, tumor_pivot.Infubinol, marker='^',color="blue", linewidth=0.5, label="Infubinol")
ketapril, = plt.plot(tumor_pivot.index, tumor_pivot.Ketapril, marker='s',color="green", linewidth=0.5, label="Ketapril")
placebo, = plt.plot(tumor_pivot.index, tumor_pivot.Placebo, marker='d',color="black", linewidth=0.5, label="Placebo")

ax.errorbar(tumor_pivot.index, tumor_pivot.Capomulin, tumor_err.Capomulin, linewidth=0.5, fmt="ro")
ax.errorbar(tumor_pivot.index, tumor_pivot.Infubinol, tumor_err.Infubinol, linewidth=0.5, fmt="b^")
ax.errorbar(tumor_pivot.index, tumor_pivot.Ketapril, tumor_err.Ketapril, linewidth=0.5, fmt="gs")
ax.errorbar(tumor_pivot.index, tumor_pivot.Placebo, tumor_err.Placebo, linewidth=0.5, fmt="kd")

# Add labels to the x and y axes
plt.title("Tumor Response To Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid(axis='y')

# Create a legend for the chart
plt.legend(handles=[capomulin, infubinol, ketapril, placebo],  loc="best")

# Save the Figure
plt.savefig("Images/treatment.png")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metastatic_group = mouse_trial_data.groupby(['Drug','Timepoint'])['Metastatic Sites'].mean().reset_index()

# Convert to DataFrame
metastatic_df = pd.DataFrame(metastatic_group)

# Preview DataFrame
metastatic_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_err_group = mouse_trial_data.groupby(['Drug','Timepoint'])['Metastatic Sites'].sem().reset_index()
# Convert to DataFrame
metastatic_std_err = pd.DataFrame(metastatic_err_group)
# Preview DataFrame
metastatic_std_err.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastatic_pivot = metastatic_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')

metastatic_err = metastatic_std_err.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
# Preview that Reformatting worked
metastatic_pivot
```

```python
metastatic_err
```

```python
# Generate the Plot (with Error Bars)
# Setting up the plot

fig, ax = plt.subplots()

capomulin, = plt.plot(metastatic_pivot.index, metastatic_pivot.Capomulin, marker='o', color="red", linewidth=0.5, label="Capomulin")
infubinol, = plt.plot(metastatic_pivot.index, metastatic_pivot.Infubinol, marker='^',color="blue", linewidth=0.5, label="Infubinol")
ketapril, = plt.plot(metastatic_pivot.index, metastatic_pivot.Ketapril, marker='s',color="green", linewidth=0.5, label="Ketapril")
placebo, = plt.plot(metastatic_pivot.index, metastatic_pivot.Placebo, marker='d',color="black", linewidth=0.5, label="Placebo")

ax.errorbar(metastatic_pivot.index, metastatic_pivot.Capomulin, metastatic_err.Capomulin, linewidth=0.5, fmt="ro")
ax.errorbar(metastatic_pivot.index, metastatic_pivot.Infubinol, metastatic_err.Infubinol, linewidth=0.5, fmt="b^")
ax.errorbar(metastatic_pivot.index, metastatic_pivot.Ketapril, metastatic_err.Ketapril, linewidth=0.5, fmt="gs")
ax.errorbar(metastatic_pivot.index, metastatic_pivot.Placebo, metastatic_err.Placebo, linewidth=0.5, fmt="kd")

# Add labels to the x and y axes
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Met. Sites")
plt.grid(axis='y')

# Create a legend for the chart
plt.legend(handles=[capomulin, infubinol, ketapril, placebo],  loc="best")

# Save the Figure
plt.savefig("Images/spread.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_group = mouse_trial_data.groupby(['Drug','Timepoint'])['Mouse ID'].count().reset_index()

# Convert to DataFrame
survival_df = pd.DataFrame(survival_group)
survival_df = survival_df.rename(columns={'Mouse ID' : 'Mouse Count'})
# Preview DataFrame
survival_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_pivot = survival_df.pivot(index='Timepoint', columns='Drug', values='Mouse Count')
# Preview the Data Frame
survival_pivot
```

```python
# Generate the Plot (Accounting for percentages)
fig, ax = plt.subplots()

capomulin, = plt.plot(survival_pivot.index, survival_pivot.Capomulin / survival_pivot.Capomulin[0] * 100, marker='o', color="red", linewidth=0.5, label="Capomulin")
infubinol, = plt.plot(survival_pivot.index, survival_pivot.Infubinol / survival_pivot.Infubinol[0] * 100, marker='^',color="blue", linewidth=0.5, label="Infubinol")
ketapril, = plt.plot(survival_pivot.index, survival_pivot.Ketapril / survival_pivot.Ketapril[0] * 100, marker='s',color="green", linewidth=0.5, label="Ketapril")
placebo, = plt.plot(survival_pivot.index, survival_pivot.Placebo / survival_pivot.Placebo[0] * 100, marker='d',color="black", linewidth=0.5, label="Placebo")

# Add labels to the x and y axes
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate %")
plt.grid(axis='y')

# Create a legend for the chart
plt.legend(handles=[capomulin, infubinol, ketapril, placebo],  loc="best")


# Save the Figure
plt.savefig("Images/survival.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
# Use the tumor pivot. The diff between last and first tumor volume over the initial tumor size.
pct_change = (tumor_pivot.iloc[-1] - tumor_pivot.iloc[0]) / tumor_pivot.iloc[0] * 100
# Display the data to confirm
pct_change
```

```python
# Store all Relevant Percent Changes into a Tuple
summary_drug_list = ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']
pct_change_list = [pct_change[x] for x in summary_drug_list]

# Splice the data between passing and failing drugs
passing_drug = pct_change < 0 # returns True for less than 0 and False for equal to or greater than 0

# Orient widths. Add labels, tick marks, etc. 

plt.grid()
plt.ylim(-30,70)
plt.ylabel('% Tumor Volume Change')
plt.title('Tumor Change over 45 Day Treatment')
#Use pandas.Series.map to map True to green and False to red color
bar_color=passing_drug.map({True:'g',False:'r'})

change_plt = plt.bar(summary_drug_list, pct_change_list, width = -1, align="edge", color=bar_color)

# Use functions to label the percentages of changes
# Reference: https://matplotlib.org/3.1.1/gallery/lines_bars_and_markers/barchart.html
def autolabel(bars):
    for i in bars:
        height = i.get_height()
        if height > 0:
            label_position = 5 # bottom left of text position
        else:
            label_position = -10 # bottom left of text position
            
        plt.text(i.get_x() + i.get_width()/2., label_position,
                str(int(height))+ '%' ,color='white', ha='center')

# Call functions to implement the function calls
autolabel(change_plt)

# Save the Figure
plt.savefig("Images/change.png")

# Show the Figure
fig.show()
```
